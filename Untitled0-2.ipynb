```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# CSV dosyasını oku
df = pd.read_csv('netflix_titles.csv')

# İlk 5 satırı göster
df.head()
```

```python
# Veri boyutu (kaç satır ve sütun)
print(df.shape)

# Kolon isimleri
print(df.columns)

# Veri tipleri ve eksik değerler
df.info()

# Kolonlardaki eksik değer sayısı
print(df.isnull().sum())
```

```python
sns.countplot(x='type', data=df)
plt.title('Film ve Dizi Sayısı')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='listed_in', data=df, order=df['listed_in'].value_counts().index)
plt.title('Tür Dağılımı')
plt.show()
```

```python
# date_added kolonunu datetime formatına çevir
df['date_added'] = pd.to_datetime(df['date_added'])

# Yıl bilgisini çıkar
df['year_added'] = df['date_added'].dt.year

# Grafik
plt.figure(figsize=(12,6))
sns.countplot(x='year_added', data=df)
plt.xticks(rotation=90)
plt.title('Yıllara Göre Netflix Yayınları')
plt.show()
```
